# Portfolio optimization

The goal of this notebook is: given a portfolio containing certain component stocks, which is the best combination that maximize return while minimizing volitility

## Import data

```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
%matplotlib inline

# Library for plotting
import matplotlib.pyplot as plt
plt.style.use('seaborn')
from matplotlib import rcParams
params = {
    'font.size': 12,
    'axes.labelsize': 15,
    'axes.titlesize': 18

}
rcParams.update(params)
rcParams['figure.figsize'] = 10, 5
rcParams['figure.dpi'] = 100
```

```python
assets = ['GOOGL','PG']
df = pd.DataFrame()
for a in assets:
    df[a] = wb.DataReader(a,data_source='yahoo',start='2015-01-01')['Adj Close']
```

```python
df.head()
```

```python
(df/df.iloc[0]).plot(figsize=(10,5))
```

```python
df_ret = df/df.shift(1)-1
df_ret.mean()*250
```

```python
df_ret.cov()*250
```

```python
df_ret.corr()
```

## Efficient frontier

```python
asset_ret = []
asset_volatile = []
w_range = np.arange(0,1.0,0.001)
for w in w_range:
    weights = np.array([w,1-w])
    asset_ret.append(np.sum(weights*df_ret.mean())*250)
    asset_volatile.append(np.sqrt(np.dot(weights.T, np.dot(df_ret.cov()*250,weights))))
    
```

```python
df_portfolio = pd.DataFrame()
df_portfolio['w'] = w_range
df_portfolio['return'] = asset_ret
df_portfolio['volatility'] = asset_volatile
```

```python
df_portfolio.head()
```

```python
fig, ax = plt.subplots()
a = plt.scatter(x=df_portfolio['volatility'], y=df_portfolio['return'], c=df_portfolio['w'], cmap='gist_rainbow')
fig.colorbar(a)
plt.xlabel('volatility')
plt.ylabel('return')
plt.title('Efficient frontier')
```

This means that one should hold at least 30% Google stock
