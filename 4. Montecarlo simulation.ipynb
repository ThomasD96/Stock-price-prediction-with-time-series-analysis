# Montecarlo simulation

This notebook aims to predict the price using Montecarlo simulation, which assumes return to follow a random variables following a normal distribution, with mean and variance obtained from historical values. As such, the future price can be largely varied from one prediction to the next

## Import libraries and data

```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
from scipy.stats import norm

# Library for plotting
import matplotlib.pyplot as plt
plt.style.use('seaborn')
from matplotlib import rcParams
params = {
    'font.size': 12,
    'axes.labelsize': 15,
    'axes.titlesize': 18

}
rcParams.update(params)
rcParams['figure.figsize'] = 10, 5
rcParams['figure.dpi'] = 100
```

```python
stock = '^GSPC'
df = pd.DataFrame()
df[stock] = wb.DataReader(stock, data_source='yahoo', start='2015-01-01')['Adj Close']
```

```python
log_ret = np.log(1+df.pct_change())
log_ret.plot(figsize=(8,5))
```

## Price prediction

```python
mean = log_ret.mean().values
var = log_ret.var().values
drift = mean-0.5*var
stdev = log_ret.std().values
```

```python
z = norm.ppf(np.random.rand(10,2))
```

```python
t_intervals = 100
iterations = 5
```

```python
daily_returns = np.exp(drift+stdev*norm.ppf(np.random.rand(t_intervals,iterations)))
```

```python
p_orig = df.iloc[-1]
p_list = np.zeros_like(daily_returns)
p_list[0] = p_orig
for i in range(1,len(p_list)):
    p_list[i] = p_list[i-1] * daily_returns[i]
```

```python
p_list = p_list.T
```

```python
start_date = df.index[-1]

fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111)
y_train = df[-300:][stock]
x_train = np.arange(-len(y_train),0,1)
plt.plot(x_train, y_train, color='black', linewidth=1.5, label='Historical price')
for i in range(len(p_list)):
    y_pred = p_list[i]
    x_pred = np.arange(1,len(y_pred)+1,1)
    plt.plot(x_pred, y_pred,label='Iteration %i'%i,alpha=0.8)

plt.legend(loc='best')
plt.xlabel('Days')
plt.ylabel('Price')
plt.title('Price prediction using Monte Carlo simulation')

# label start date
fig.canvas.draw()
labels = [item.get_text() for item in ax.get_xticklabels()]

for i in range(len(labels)):
    if labels[i] == '0':
        labels[i] = start_date
ax.set_xticklabels(labels)
```
