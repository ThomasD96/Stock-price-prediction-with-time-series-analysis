```python
rev_m = 170
rev_std = 20
iterations = 1000
```

```python
# cogs = cost of goods sold
cogs = - (rev*np.random.normal(0.6,0.1))

plt.figure(figsize=(10,5))
plt.plot(cogs)
plt.xlabel('Iteration')
plt.ylabel('Cogs')
plt.show()
```
