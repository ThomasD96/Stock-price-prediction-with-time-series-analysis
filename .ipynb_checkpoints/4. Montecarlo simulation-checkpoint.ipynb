# Montecarlo simulation

This notebook aims to predict the price using Montecarlo simulation, which assumes price change as a random variables following a normal distribution, with mean and variance obtained from historical values. As such, the future price can be largely varied from one prediction to the next

## Import libraries and data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas_datareader import data as wb
from scipy.stats import norm
```

```python
stock = '^GSPC'
df = pd.DataFrame()
df[stock] = wb.DataReader(stock, data_source='yahoo', start='2015-01-01')['Adj Close']
```

```python
log_ret = np.log(1+df.pct_change())
log_ret.plot(figsize=(8,5))
```

## Price prediction

```python
mean = log_ret.mean().values
var = log_ret.var().values
drift = mean-0.5*var
stdev = log_ret.std().values
```

```python
z = norm.ppf(np.random.rand(10,2))
```

```python
t_intervals = 1000
iterations = 10
```

```python
daily_returns = np.exp(drift+stdev*norm.ppf(np.random.rand(t_intervals,iterations)))
```

```python
daily_returns
```

```python
p_orig = df.iloc[-1]
p_list = np.zeros_like(daily_returns)
p_list[0] = p_orig
for i in range(1,len(p_list)):
    p_list[i] = p_list[i-1] * daily_returns[i]
```

```python
plt.figure(figsize=(10,6))
plt.plot(p_list)
```
